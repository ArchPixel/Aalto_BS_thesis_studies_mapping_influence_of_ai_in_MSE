```python
import pandas as pd

df_docs = pd.DataFrame(columns=['DOI','Title','Abstract', 'Year'])
for item in docs_ai:
    row = pd.DataFrame({'DOI':[item.id.doi], 'Title':[item.title], 'Abstract': [item.abstract], 'Year': [item.publication_year]})
    df_docs = pd.concat([df_docs, row], ignore_index = True)

print(df_docs)
rac_docs_ai_csv = df_docs.to_csv('rac_docs_ai.csv', index=True)
print('\nCSV String:\n', rac_docs_ai_csv)
print

```
